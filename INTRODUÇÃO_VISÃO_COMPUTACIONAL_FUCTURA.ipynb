```
image_carro = image_original.copy()
cv2.line(image_carro, (10, 10), (100, 100), (0, 255, 0), 5)
cv2_imshow(image_carro)
```

```
image_carro = image_original.copy()
cv2.circle(image_carro, (int(image_carro.shape[1]/2), int(image_carro.shape[0]/2)), 50,(20, 60, 40), -1)
cv2_imshow(image_carro)
```

```
image_carro = image_original.copy()
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(image_carro, 'Curso de Python III - FUCTURA', (50, 50), font, 0.5, (0, 0, 255), 1, cv2.LINE_AA) 
cv2_imshow(image_carro)
```

```
image_carro = image_original.copy()
image_carro = cv2.cvtColor(image_carro, cv2.COLOR_RGB2GRAY)
cv2_imshow(image_carro)
```

```
ret, thresh1 = cv2.threshold(image_carro,100,300,cv2.THRESH_BINARY)
cv2_imshow(thresh1)
```

```
image_carro = cv2.cvtColor(thresh1, cv2.COLOR_GRAY2RGB)
image_carro[thresh1 == 255] = (255,0,0)
image_carro[thresh1 == 0] = (0,255,0)

cv2_imshow(image_carro)
```

```
from matplotlib import pyplot as plt
hist = cv2.calcHist([thresh1],[0],None,[256],[0,256])
plt.hist(thresh1.ravel(),256,[0,256])
plt.show()
```

```
ret, thresh1 = cv2.threshold(r,140,255,cv2.THRESH_BINARY)
cv2_imshow(thresh1)
```
